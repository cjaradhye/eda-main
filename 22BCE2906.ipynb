```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('https://raw.githubusercontent.com/salemprakash/EDA/main/Data/Fertility2.csv')
```

```python
# Ex. No. 7 (a): 1-Dimensional Analysis

print("1-Dimensional Analysis:")
print("\nSummary Statistics:")
print(data.describe())

print("\nValue Counts for Categorical Variables:")
for col in ['morekids', 'gender1', 'gender2', 'afam', 'hispanic', 'other']:
    print(f"\n{col}:")
    print(data[col].value_counts())
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='age', kde=True)
plt.title('22BCE2906 - Age Distribution')
plt.tight_layout()
plt.show()
```

```python
# Work hours distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='work', kde=True)
plt.title('22BCE2906 - Work Hours Distribution')
plt.tight_layout()
plt.show()
```

```python
# Work hours distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='work', kde=True)
plt.title('22BCE2906 - Work Hours Distribution')
plt.tight_layout()
plt.show()
```

```python
# Gender1 distribution
plt.figure(figsize=(10, 6))
data['gender1'].value_counts().plot(kind='bar')
plt.title('22BCE2906 - Gender1 Distribution')
plt.tight_layout()
plt.show()
```

```python
# Morekids distribution
plt.figure(figsize=(10, 6))
data['morekids'].value_counts().plot(kind='bar')
plt.title('22BCE2906 - More Kids Distribution')
plt.tight_layout()
plt.show()
```

```python
# Ex. No. 7 (b)

# Correlation matrix
correlation_matrix = data[['age', 'work']].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# Box plot: Work Hours by Gender1
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='gender1', y='work')
plt.title('22BCE2906 - Work Hours by Gender1')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Age vs Work Hours
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='age', y='work', hue='gender1')
plt.title('22BCE2906 - Age vs Work Hours')
plt.tight_layout()
plt.show()
```

```python
# Bar plot: Average Age by Morekids
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='morekids', y='age')
plt.title('22BCE2906 - Average Age by More Kids')
plt.tight_layout()
plt.show()
```

```python
# Heatmap of correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('22BCE2906 - Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python

# Additional N-D Analysis: Pairplot
pairplot = sns.pairplot(data, vars=['age', 'work'], hue='gender1')
pairplot.fig.suptitle('22BCE2906 - Pairplot: Age, Work Hours, and Gender1', y=1.02)
plt.show()
```
